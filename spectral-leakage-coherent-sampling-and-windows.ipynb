# Hello there

This is going to be some plain text. This is going to be *LATEX*. $2x^{2_1}$

```julia
f_s = 1e6;
T_s = 1/f_s;

n_fft = 2^12;
k = 11;

f_i = f_s/n_fft*k;

t = 0:T_s:T_s*(n_fft-1);
i = sin.(2*pi*f_i*t);

PyPlot.plot(t, i)
```

```julia
Y_i = fft(i)/n_fft;

f = linspace(0, f_s/2, (n_fft/2+1));

PyPlot.semilogx(f[2:end], 20*log10.(2*abs.(Y_i[2:trunc(Int64, n_fft/2+1)])))
```

```julia
f_s = 1e6;
T_s = 1/f_s;

n_fft = 2^16;

f_i = rand()*f_s/4;

t = 0:T_s:T_s*(n_fft-1);
i = sin.(2*pi*f_i*t);

Y_i = fft(i)/n_fft;

f = linspace(0, f_s/2, (n_fft/2+1));

PyPlot.semilogx(f[2:end], 20*log10.(2*abs.(Y_i[2:trunc(Int64, n_fft/2+1)])));
```
